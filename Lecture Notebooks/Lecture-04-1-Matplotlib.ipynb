## Basic plotting

The basic procedure to plot a function $y=f(x)$ is:

1. Create an array of `x` values, for example using `numpy.linspace`.
2. Create an array of `y` values using array operations and functions.
3. Use `plt.plot(x,y)`.
4. Add style using `pyplot` commands.
5. Show the plot with `plt.show()`
6. If desired, save the plot using `plt.savefig('filename', options)`

For example, let's plot $y = x^3$ for $x$ between 0 and 3

```python
x = np.linspace(0,3,100)
y = x**3
plt.plot(x,y)
plt.show()
```

You can also plot data, with scatter points and/or with a line

```python
x = np.array([2.1,3.5,4.2,5.1,6.0])
y = np.array([0.2,0.1,0.05,0.03,0.025])
plt.plot(x,y,'ro')  # using red circle markers
#plt.scatter(x,y) #another option for scatter plots
plt.show()
```

A plot can show multiple data sets. The example below shows two, 
but one can plot together as many data sets as desired

```python
x1 = np.linspace(0,3,100)
y1 = x1**3

x2 = np.linspace(0,4,100)
y2 = 2*x2
plt.plot(x1,y1) 
plt.plot(x2,y2) # works for as many data sets as desired
# alternative implementation:
# plt.plot(x1,y1,x2,y2)
plt.show()
```

### Saving the plot

## Customizing plots

### Axes, grids, legend, ticks, text

We can use `pyplot` options to customize your plot. For example, you can add a plot title, axes labels, plot legend, and grids.<br>

```python
x = np.linspace(0,3,100)
y = x**3
plt.plot(x,y)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('my plot')

plt.grid(True)
plt.show()
```

There are countless options and consulting the matplotlib website or online examples and galleries is very helpful.

Let's explore a few more basic options.<br>
Use `xlim` and `ylim` to set the axes limits, and `xticks` and `yticks` to set the axes ticks.<br> 
Use `plot.legend()` to create a legend.<br>
Use `label` to create a label for the legend.<br>
You can use Latex in the plot using `r'$math$'`
The font size can be modified using the fontsize command.


```python
x = np.linspace(0,10,100) # x is defined up to 10
y = x**3

plt.xlim(0,5)
plt.ylim(0,150)

# if setting ticks, the limits need to be consistent with xlim, ylim
# xticks and yticks take a list or tuple or numpy array
plt.xticks([0,2.5,5])
plt.yticks(np.arange(0,150.1,30)) 

plt.plot(x,y, label=r'$y=x^3$') # add a label to y using latex

plt.legend(loc='upper left',fontsize=22) #shows the legend in the upper left
plt.show()
```

### Lines, markers, colors, error bars, annotations

A line appearing in a plot has several properties including width, color, and style.<br>
One can also show markers, with or without a line in addition to the markers.<br>
We can set line and marker properties using keyword arguments.<br>
Here are some useful ones:

`linestyle` (or `ls`) can be `solid`, `dashed`, `dashdot`, `dotted`<br>

`linewidth` (or `lw`) can be set to a number<br>

`color` (or `c`) can be set to any color (`red`, `green`, `blue`), or to an RGB color using a tuple such as (0,1,1)<br>

`marker` sets the marker style<br>

`markersize` (or `ms`) can be set to a number<br>

`markerfacecolor` (or `mfc`) and `markeredgecolor` (or `mec`) can be set to any color

Here are some examples that combine these properties:<br>

```python
x = np.linspace(-2,2,41)
y = np.exp(-x**2) * np.cos(2*np.pi*x)

# use both blue markers and red dashed line
# note the formula written with a Latex command
plt.plot(x,y,label=r'$e^{-x^2} \cos(2\pi x)}$',
         color='red',linestyle='dashed',linewidth=2, 
         marker='o',markersize=5,markerfacecolor='blue',
         markeredgecolor='blue')

plt.ylim([-1.5,2])
plt.legend(loc='upper center', fontsize=16)
plt.show()
```

It is common to use the abbreviated version of these commands:
1. For colors, `b`=blue, `r`=red, `k`=black, `c`=cyan, `m`=magenta, etc
2. For lines, `-`=solid line, `--`=dashed line, `-.`=dash dot line, `:` dotted line
3. For markers, `o`=circle, `.`=point, `D`=diamond, `^`=triangle, `s`=square, etc

```python
# same plot as above
plt.plot(x,y,'r--',marker='o', mfc='b', mec='b',ms=5,lw=2,label=r'$e^{-x^2} \cos(2\pi x)}$')

plt.ylim([-1.5,2])
plt.legend(loc='upper center', fontsize=16)
plt.show()
```

We can also add error bars to the plots (their value can be read from file or defined in python)

```python
x = np.arange(0, 4, 0.2) 
y = np.exp(-x) 

# error bars on y and x are numpy arrays the size of x and y
dy = 0.1 * np.abs(np.random.randn(len(y))) # a random value
dx = 0.2 * np.ones_like(y) # a fixed value 

# print out the errors
print (dy,'\n')
print (dx)

plt.errorbar(x,y, xerr=dx,yerr=dy, marker='o',markersize=8.0, capsize=0.0,elinewidth=1.0);
```

We can annotate the plots with text, arrows, lines, etc

```python
# Generate a linear plot and show the intercept
    
x = np.linspace(0, 0.1, 100)
yint = 0.15
y = yint + 2.0*x

plt.plot(x,y, marker='None', linestyle='-')
plt.axis([0,0.09,0,0.4])
plt.xlabel('x',fontsize=18)
plt.ylabel('f(x)',fontsize=18)

# add text at position (0.002,0.20) 
plt.text(0.002,0.25, r'$f(0) = 0.15$', fontsize=18)
    
# add text at a location expressed in figure coordinates
plt.figtext(0.2,0.2,'intercept = 0.15', fontsize=18)
    
# add an arrow: 
# arrow(x, y, dx, dy, **kwargs) draws an arrow from (x, y) to (x+dx, y+dy).
plt.arrow(0.01,0.08, -0.0073,0.06, facecolor='m', edgecolor='m')

plt.show();
```

### Subplots

We can create subplots stacked vertically and/or horizontally, or arranged in any desired grid.

```python
# stack along the y axis: (21#) = (2x1 arrangement, plot #)

import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k-')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.tight_layout()
plt.show()
```

## Plot types and styles

Plot axes can be linear or logarithmic.<br>
Properties that apply to the entire plot can be set using plt.rcParams; for a full list, type plt.rcParams.keys()<br>

Many types of plots are possible, including polar plots, histograms, scatter plots, bar charts and pie charts.<br>
We show a few examples below; for a complete list of available plot types, see http://matplotlib.org/gallery.html.

### Log plots

```python
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams['font.size'] = 16

# make up some data in the interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(311)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(312)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# log-log
plt.subplot(313)
plt.plot(x, y)
plt.yscale('log')
plt.xscale('log')
plt.title('log-log')
plt.grid(True)


plt.tight_layout() #usually results in better looking plots
plt.show()
```

### Polar plot

```python
plt.rcParams['figure.figsize'] = [4, 4]

# polar plot 
theta = np.linspace(0, 2.0*np.pi, 100) # the angle between 0 and 2pi, with 100 intervals
plt.polar(3*theta, theta/5, lw=3, c='m');
```

### Histogram

```python
# get 10000 gaussian distributed random numbers
h = np.random.randn(10000)

# the number of bins in our histogram; the keyword density normalizes the distribution to 1
numBins = 10
plt.hist(h,numBins,edgecolor='k', color='g',density=True, alpha=0.6) # alpha is the transparency

# let's add the theoretical value
x = np.arange(-5,5,0.01)
y = np.exp(-x**2/2)/np.sqrt(2.0*np.pi)
plt.plot(x,y,'b-');

# we can save our figure to disk
plt.savefig('histogram.pdf')
```

## "Object-oriented" Matplotlib 

We have made use of the functions implemented in `pyplot`.<br> 
These functions are simple to use; internally, they make use of two main objects that represent a plot, the `figure` and `axes` objects.<br>
For better control of our plots, we can act directly on the `figure` and `axes` objects using appropriate methods, 
but first we need to understand what they are:
1. Figure: The Figure is the overall window or page where everything is drawn. It’s the top-level component. You can add text, a legend, etc. to the figure.
2. Axes: to the figure you add axes. The Axes is the area on which the data is plotted with functions such as `plot()` and `scatter()`, and that can have ticks, labels, etc. associated with it. Figures can contain multiple Axes.

For example, when you use `plt.xlim`, you are in fact calling ax.set_xlim() behind the scene, a function that acts on the object `ax` representing the axes. This approach is particularly useful when you have multiple subplots.<br>

If you want to get a deeper understanding of this approach to using matplotlib, this is an [excellent tutorial](https://www.datacamp.com/community/tutorials/matplotlib-tutorial-python)<br>
Let's see a simple example with just one plot:

```python
plt.rcParams['figure.figsize'] = [4, 4]
plt.rcParams['font.size'] = 10
# Create the Figure object
fig = plt.figure()

# Set up the Axes object as ax (there is only one subplot in this case)
ax = fig.add_subplot(111) 

# another way of creating the axes is adding them to the figure
#ax = fig.add_axes([0,0,1,1]) #lower left point is (0,0), and width and height are set to 1

# Scatter the data
ax.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

# Show the plot
plt.show()
```

```python
# yet another way is the following, which is widely used in online tutorials:
fig, ax = plt.subplots()

x = np.linspace(0, 1, 500)
y = np.sin(4 * np.pi * x) * np.exp(-5 * x)

ax.plot(x, y)
plt.show()
```

For subplots, using figures and axes is particularly useful, and in fact most examples on the Matplotlib website<br> will follow the Figure / Axes plus subplots object oriented approach. Let's see an example:

```python
fig = plt.figure()
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

# Plot the data
ax1.bar([1,2,3],[3,4,5]) #first subplot is a bar chart
ax2.barh([0.5,1,2.5],[0,1,2]) #second subplot is a horizontal bar chart
ax3.scatter(x,y) # third subplot is a scatter plot

# Show the plot
plt.tight_layout()
plt.show()
```

Using these methods, we can create a plot with an inset

```python
plt.rcParams['font.size']=18
fig = plt.figure()

x = np.linspace(0, 5, 10)
y = x ** 3

ax1 = fig.add_axes([0, 0, 2.0, 1.0]) # main axes
ax2 = fig.add_axes([0.2, 0.6, 0.6, 0.3]) # inset axes

# main figure
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# inset
ax2.plot(y, x, 'g')
ax2.set_xlabel('y')
ax2.set_ylabel('x')
ax2.set_title('my inset');
```
